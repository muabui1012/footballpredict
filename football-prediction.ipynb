## Get data

```python
import json
import pandas as pd

import aiohttp

from understat import Understat
import json


async with aiohttp.ClientSession() as session:
    understat = Understat(session)
    fixtures =  await understat.get_league_results("epl",2023)
    print(fixtures[:2])
```

## Clean data

```python
import pandas as pd

result_dict_list = []

for item in fixtures:
  result_dict = {
    'home_team': item['h']['title'],
    'away_teaam': item['a']['title'],
    'home_goals': item['goals']['h'],
    'away_goals': item['goals']['a'],
    'home_xG': item['xG']['h'],
    'away_xG': item['xG']['a'],
  }
  result_dict_list.append(result_dict)

pd.DataFrame(result_dict_list)
```
